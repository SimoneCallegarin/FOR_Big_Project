## Biogas plants location

An association of $n$ farmers wants to open $p$ plants to produce energy from biogas. 
Each plant will be opened at a farm of a member of the association and will be powered with corn chopping purchased from the farm itself or from other neighboring farms.

Each farm $i$ can provide at most $c_i$ tons of corn chopping, with a percentage of dry matter $a_i$. As you may know, dry matter is the key component of corn chopping used for biogas production. In order to maintain the quality of produced energy, each plant must burn a mixture of corn chopping with a percentage of dry matter between $k_{min}$ and $k_{max}$. 

At most one plant can be located in each farm, and every farm can sell its corn chopping to one and only one plant.

Each farm $i$ is located at coordinates $x_i$ and $y_i$, representing respectively its latitude and longitude, and the cost of moving corn chopping from a farm $i$ to a farm $j$ is proportional to the euclidean distance between the two farms (it does not depend on the actual quantity moved, since the trucks used for this transportations are sufficiently big). 

Under such conditions, every plant produces $Q$ kWh of energy per ton of corn chopping burned. The energy produced by each plant will be fed into the national electricity system, at a unitary price of $b$ (€/kWh). Moreover, due to state regulations, each plant must not produce more than $M$ kWh of energy.

You must locate $p$ plants among the available farms and assign the farms that will supply each plant, with the goal of maximizing the total revenues of the association.

### Sets
*   $I$ = set of farms

### Parameters
*   $n$ = number of farms   
*   $p$ = number of plants to locate
*   $b$ = revenue per unit of energy (€/kWh)
*   $M$ = max energy production (kWh)
*   $Q$ = energy produced by a ton of corn chopping (kWh/t)
*   $k_{min} (k_{max})$ = min (max) percentage of dry matter for fermentation
*   $a_i$ = percentage of dry matter in chopping from farm $i \in I$
*   $c_i$ = tons of corn chopping available for each $i \in I$ (t)
*   $x_i, y_i$ = coordinates of farm $i \in I$

### Other functions
Some other useful function that are called by the solve function.

#### Euclidean distances
Computes the euclidean distances between all pairs of point of the instance.
Then returns a list of lists containing all these values.

In order to accomplish this faster we exploited the "Euclidean Trick".

Documentation here:


*   https://www.robots.ox.ac.uk/~albanie/notes/Euclidean_distance_trick.pdf
*   https://github.com/eth-cscs/PythonHPC/blob/master/numpy/03-euclidean-distance-matrix-numpy.ipynb
*   https://github.com/Daddeee/FOR_Labs_22-23/blob/master/lab2/solutions/lab2-shortest_path.ipynb



```python
# Computes the Euclidean Distances between each pair of points
def fast_distance_matrix(x):
    xy = x @ x.T
    x2 = xy.diagonal()[:,np.newaxis]
    return np.abs(x2 + x2.T - 2. * xy)**0.5
```

#### Our idea
Next are two function that are the core of our solution.

Our idea relies on observations made from the ten instances that were provided to us.
After some reasoning we came to the conclusion that the gain coming from the burn of some material weights far times more than the cost of transport (i.e. the euclidean distance between a farm and a plant that buys from the first one).

We extimated that the revenues are more than a hundred times __per tons__ bigger than the costs:
$ ( \sum_{i\in N} \sum_{j\in N} xb_{ij} ) ⋅ Q ⋅ b >> \sum_{i\in N} \sum_{j\in N} (xfp_{ij} \cdot distance\_matrix_{ij}) $ 

and considering that each farm in each instance have on average around 1000 tons of corn choppings available and that plants are able to buy from many different farms (we saw that usually the number varies from from 1 to dozens of farms) the math is soon done.

Infact we noticed that distances reach on average values of around 70 that are small fries compared to a plant burning far more than 1000 tons of material that generates revenue of 100 per ton.

These are the reasons that brought us to develop the following solution focusing on burn the highest amount of corn choppings and on second hand on the distances.



##### Locate plants $xp$
Called after the first optimization of the model with fractional values.

It searches for the maximum fractional values between the $xp$ of the farms and then proceeds to store them in a list $plants$ that will contain the keys of the farms that will most likely be plants (because they burn the higher quantity of material in the first model optimization).

It repeats the procedure for $p$ times in order to put in the list the $p$ farms that will probably be the best options to become plants.

```python
# Proceed to put in the list the farms that are the ones most likely to be plants and returns the list:
def locate_plants(xp):

    plants = []

    for i in range(p):

        # Search for the maximum value of xp[key].x in xp that is not already in plants
        max_key = max(xp, key=lambda key: xp[key].x and key not in plants)

        # Store the key of the plant in plants
        plants.append(max_key)
    
    return plants
```

##### Cutting function on $xp$
With this function we are trying to add cuts to the first optimal solution found on the previous model.

---

Has previously seen, the idea is to identify the $p$ plants with the highest fractional values and store them in the list $plants$(exploiting the previous function).

Then for all the farms that are not in $plants$ (that still have fractional values of $xp$) we impose a constraint on their $xfp$ that forces the $sum$ of their $xfp$ to be lesser or equal then 0 (it will obviously assumes value of 0 due to the fact that the lower bound $lb$ for $xfp$ is set to 0).

---

The function follows these steps in order to accomplish that:
1.   Puts the $p$ farms that are most likely the ones to be plants in the list $plants$.
2.   Loops on $i$ for every farm.
3.   If the farm is not in the list $plants$ (so the farms that are not considered as plants) it adds a constraint over these farms' $xfp$ in order to set them all to 0.
4.   Calls $produce\_energy $ function (whose operation is explained below) in order to check that the constraint on M has not been violated after the adjust on xp.


```python
# Adjust xp by putting constraints on the farms that are probably not plants:
def adjust_xp(xb,xp,xfp,m):

    # Locate plants
    plants = locate_plants(xp)

    for i in range(N):
      if i not in plants:
        m.add_constr( mip.xsum( xfp[i,j] for j in range(N) ) <= 0 ) 

    m.optimize()

    # Check if the constrain on M has been violated and proceeds to correct it
    produce_energy(xb,xp,plants,m)
```

#### Constraint on M
Each plant must not produce more than  M  kWh of energy.

We call this function after the first optimization of the model and then after the adjust on xp in order to check if the constraint on the M has been violated by some plants.

This actually helps quiet a lot for the time optimization due to the fact that it's a constraint less for the model's optimization and that is applyed only further on and on the farms that are already known to be plants (so many more than the total number of farms $n$).

```python
# Puts the constarint on the maximum quantity of energy (M) produced by each plant:
def produce_energy(xb,xp,plants_xp,m):

    #print(f"new objective: {m.objective_value}")
  
    for i in plants_xp:
      
        if sum( xb[i,j].x for j in range(N) ) * Q > M * xp[i].x:

            m.add_constr(mip.xsum( xb[i,j] for j in range(N) ) * Q <= M * xp[i] )   

    m.optimize()
```

### Solve function
This is the function that defines the model with its variables, constraints and the objective function.

---

#### Variables

*   $xb$: Quantity of corn burned by each farm that was bought from a certain farm
*   $xp$: States if a farm is also a plant
*   $xfp$: Indicates if a farm j sells to a plant i

---

#### Constraints

*   Number of plant must be equal to $p$
*   Every farm can sell its corn chopping to one and only one plant
*   Every plant can buy from a farm only a quantity that is less than the availabile quantity of the farm
*   Each plant must burn a mixture of corn chopping with a percentage of dry matter between  $kmin$  and  $kmax$
*   Each farm buying corn chopping is a plant
*   Each plant must not produce more than  M  kWh of energy (as previously seen this constraint is added only at a certain time during the solve function)

---

#### Objective function
Objective function = Revenues - Costs

*   Revenues = $ ( \sum_{i\in N} \sum_{j\in N} xb_{ij} ) ⋅ Q ⋅ b $
*   Costs =  $ \sum_{i\in N} \sum_{j\in N} (xfp_{ij} \cdot distance\_matrix_{ij}) \to $  where $distance\_matrix_{ij}$ is the evaluated distance between farm $i$ and $j$

---

#### Optimize
We call for the first time m.optimize() and then we call the functions explained above in order to find the plants and correct violated constraints.

In the end the function returns the objective value of the objective function.

```python
# Solves the model:
def solve(instance):

  m = mip.Model()

# VARIABLES (integer variables are relaxed into continuous variables):

  xb = { (i,j) : m.add_var(lb=0, ub=c[j]) for j in range(N) for i in range(N) } # Quantity of corn burned by each farm that was bought from a certain farm
  xp = { i : m.add_var(lb=0, ub=1) for i in range(N) } # Indicates if a farm is also a plant
  xfp = { (i,j) : m.add_var(lb=0, ub=1) for j in range(N) for i in range(N) } # Variable saying if a farm j sells to a plant i


# CONSTRAINTS:

  # The number of plant must be equal to p:
  m.add_constr( mip.xsum( xp[i] for i in range(N) ) == p ) 

  # Every farm can sell its corn chopping to one and only one plant:
  for i in range(N):
    m.add_constr( mip.xsum( xfp[j,i] for j in range(N) ) == 1 ) 

  # Every plant can buy from a farm only a quantity that is less than the availabile quantity of the farm:
  for i in range(N):  
    for j in range(N):
      m.add_constr( xb[i,j] <= c[j] * xfp[i,j] )

  # Each plant must burn a mixture of corn chopping with a percentage of dry matter between  kmin  and  kmax:
  for i in range(N):  
    m.add_constr( mip.xsum( xb[i,j] * a[j] for j in range(N) ) <= kmax * mip.xsum( xb[i,j] for j in range(N) ) ) 
    m.add_constr( mip.xsum( xb[i,j] * a[j] for j in range(N) ) >= kmin * mip.xsum( xb[i,j] for j in range(N) ) )

  # Stating that each farm buying corn chopping is a plant:
  for i in range(N):
    m.add_constr( mip.xsum( xfp[i,j] for j in range(N) ) >= xp[i] )
    

  # Computing the distances between each pair of points:
  distance_matrix = fast_distance_matrix( np.array( points ) )


# OBJECTIVE FUNCTION (revenues - costs):
  m.objective = mip.maximize(  mip.xsum( xb[i,j] for j in range(N) for i in range(N) ) * Q * b
                             - mip.xsum( xfp[i,j] * distance_matrix[i][j] for j in range(N) for i in range(N) ) 
                             ) 

# OPTIMIZE: 
  m.optimize()

  produce_energy(xb,xp,xp,m)
  
  # Adjusting the fractional values of xp in order to reach the solution:
  adjust_xp(xb,xp,xfp,m)

  return m.objective_value
```
